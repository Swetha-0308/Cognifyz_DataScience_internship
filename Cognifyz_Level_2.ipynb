#Table Booking and Online Delivery( Task 1 )

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('/content/Dataset .csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
table_booking = data['Has Table booking'].value_counts()
print(table_booking)
```

```python
has_table_booking = data[data['Has Table booking']=="Yes"]
has_table_booking
```

```python
has_table_booking.shape
```

```python
online_delivery = data['Has Online delivery'].value_counts()
print(online_delivery)
```

```python
has_online_delivery = data[data['Has Online delivery']=="Yes"]
has_online_delivery
```

```python
has_online_delivery.shape
```

```python
Has_table_booking = len(has_table_booking)
Has_online_dekivery = len(has_online_delivery)
```

```python
table_booking_per = (Has_table_booking/len(data))*100
online_delivery_per = (Has_online_dekivery/len(data))*100

print(f"percentage of restaurants that offer table Booking : {table_booking_per:2f}%")
print('\n')
print(f"percentage of restaurants that offer online delivery : {online_delivery_per:2f}%")
```

```python
percentages = [table_booking_per ,100-table_booking_per, online_delivery_per, 100-online_delivery_per]
labels = ['Table Booking', 'Not Table Booking', 'Online Delivery', 'Not Online Delivery']
colors = ['blue', 'yellow', 'blue', 'yellow']
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.pie(percentages[:2], labels= labels[:2], colors=colors[:2], autopct='%1.1f%%', startangle=90)
plt.title('Table Booking')

plt.subplot(1,2,2)
plt.pie(percentages[2:], labels= labels[2:], colors=colors[2:], autopct='%1.1f%%', startangle=90)
plt.title('Online Delivery')

plt.tight_layout()
plt.show()
```

```python
data.loc[:, ['Has Table booking', 'Aggregate rating']]
```

```python
has_table_booking_mean= data[data['Has Table booking']=="Yes"]['Aggregate rating'].mean()
not_have_table_booking_mean= data[data['Has Table booking']=="No"]['Aggregate rating'].mean()

print(f"Average rate of Restaurants that offer table booking : {has_table_booking_mean:2f}%")
print('\n')
print(f"Average rate of Restaurants that does not offer table booking : {not_have_table_booking_mean:2f}%")
```

```python
import matplotlib.pyplot as plt

#data
category=['with Table Booking ',' Without table Booking' ]
average_ratings=[has_table_booking_mean, not_have_table_booking_mean]

#plot
plt.figure(figsize=(8,10))
plt.bar(category, average_ratings, color=['pink', 'lavender'])
plt.title("Average rate of restaurants vs Table Booking")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.ylim(0,5)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
data.loc[:, ['Has Online delivery','Price range']]
```

```python
data['Restaurant Name'].value_counts().shape
```

```python
online_delivery =data.groupby('Price range')['Has Online delivery'].value_counts().unstack()
print(online_delivery)
```

```python
#poltting
plt.figure(figsize=(8,6))
online_delivery.plot(kind='bar', stacked=True, color=['pink', 'lavender'])
plt.title('Online Delivery Availability by Price Range')
plt.xlabel('Price Range')
plt.ylabel('No.of restaurants')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Online Delivery')
plt.show()
```

#Price Range Analysis( Task 2)

```python
price_range = data['Price range'].value_counts()
price_range.plot(kind='bar', color='pink')
plt.xlabel('Price Range')
plt.ylabel('No.of restaurants')
plt.title('Price Range Analysis')
plt.grid(axis='y',linestyle='--', alpha=0.7)
plt.show()

common_price_range = price_range.idxmax()
print("common price range among all restaurants: ", common_price_range)

```

```python
rating_mean= data.groupby('Price range')['Aggregate rating'].mean()
print(rating_mean)
```

```python
rating_mean.plot(kind='bar', color='skyblue')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.title('Price Range vs Average Rating')
plt.grid(axis='y',linestyle='--', alpha=0.7)
plt.show()
```

```python
high_rating_mean= rating_mean.idxmax()
plt.bar(rating_mean.index, rating_mean, color='lavender')
plt.bar(high_rating_mean, rating_mean[high_rating_mean], color='pink')
plt.title("Highest average rating by price range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.show()
print('\n')
print("Highest Avarege rating among different price ranges color :", "pink")
```

Handling the missing values

```python
missing_values = data.isnull().sum()
print(missing_values)
print("\n")
print("columns with missing values: ")
print(missing_values[missing_values>0])
```

```python
counts=data['Aggregate rating'].value_counts()
print("counts: ")
print(counts)
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.countplot(x='Aggregate rating',data=data,palette="Set3",order=data['Aggregate rating'].value_counts().index)
plt.xlabel('Aggregate rating')
plt.ylabel('Count')
plt.title('Distribution of Aggregate rating')
plt.show()
```

```python
target_variable = data['Aggregate rating'].value_counts().sort_index()
target_variable
class_imb = target_variable/target_variable.sum()
class_imb
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Aggregate rating', data=data)
plt.title("Box plot")
plt.xlabel("Aggregate rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Aggregate rating'], kde=True, color='skyblue')
plt.title("Histogram")
plt.xlabel("Aggregate rating")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#set the figure size for better visualization
plt.figure(figsize=(8,8))

#plot the distribution of cuisines(showing top 10 cuisines for bettter visualization)
top_cuisines = data['Cuisines'].value_counts().nlargest(10)
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="Set3")
plt.title('top 10 cuisines')
plt.xlabel('Count')
plt.show()


```

```python
top_cities = data['City'].value_counts().head(10)
top_cuisines = data['Cuisines'].value_counts().head(10)
print("Top 10 cities with the highest number of restaurants:")
print(top_cities)
print("\n")
print("Top 10 cuisines with the highest number of restaurants:")
print(top_cuisines)
```

# **Geospatial Analysis( Task 3 )**

```python
import plotly.express as px
import pandas as pd

fig=px.scatter_mapbox(data, lat= "Latitude", lon="Longitude", hover_name="Restaurant Name", hover_data= ["City","Cuisines"])

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
country_res_dist = data['Country Code'].value_counts()
country_res_dist
```

```python
city_res_dist = data['City'].value_counts()
city_res_dist
```

```python
fig= px.scatter(data,x="Longitude", y="Latitude", color="Aggregate rating", hover_name="Restaurant Name", color_continuous_scale='viridis')
fig.update_layout(title="Scatter plot vs Aggregate rating", xaxis_title="Longitude", yaxis_title="Latitude")
fig.show()
```

```python
plt.figure(figsize=(8,6))
correlation_matrix = data[["Latitude", "Longitude", "Aggregate rating"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Restaurant's location and rating")
plt.show()
```
